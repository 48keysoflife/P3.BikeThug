```python
bike_data_grouped = bike_data.groupby(['year', 'month']).size().reset_index(name='Theft_Count')

# Create a new 'date' column representing the first day of each month for plotting
bike_data_grouped['date'] = pd.to_datetime(bike_data_grouped[['year', 'month']].assign(day=1))

# Define season colors
def get_season_color(month):
    if month in [3, 4, 5]:      # Spring
        return 'green'
    elif month in [6, 7, 8]:    # Summer
        return 'orange'
    elif month in [9, 10, 11]:  # Autumn
        return 'brown'
    else:                       # Winter
        return 'blue'

# Apply season colors
bike_data_grouped['Season_Color'] = bike_data_grouped['month'].apply(get_season_color)

# Plot
plt.figure(figsize=(20, 8))
plt.plot(bike_data_grouped['date'], bike_data_grouped['Theft_Count'], marker='o', color='grey', alpha=0.6, linestyle='--')

# Scatter plot with color-coded seasons
for season, color in zip(['Spring', 'Summer', 'Autumn', 'Winter'], ['green', 'orange', 'brown', 'blue']):
    subset = bike_data_grouped[bike_data_grouped['Season_Color'] == color]
    plt.scatter(subset['date'], subset['Theft_Count'], color=color, label=season, s=80)

# Customize plot
plt.title('Bike Theft Incidents by Month (2014-2024)', fontsize=16)
plt.xlabel('Months', fontsize=12)
plt.ylabel('Number of Bike Thefts', fontsize=12)

# Format x-axis labels as months without leading zeros
plt.xticks(bike_data_grouped['date'], bike_data_grouped['month'], rotation = 45)

plt.legend(title='Season', fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.ylim(0, 700)
# Show the plot
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
bike_data['Theft_Risk'] = bike_data['STATUS'].apply(lambda x: 1 if x == 'STOLEN' else 0)

# Select features
features = ['LOCATION_TYPE', 'BIKE_COST', 'month', 'REPORT_HOUR','day_of_the_week','x','y']  # Include OCC_HOUR as 'hour'
bike_data = bike_data.dropna(subset=['BIKE_COST'])

# Encode the categorical variable 'LOCATION_TYPE'
le = LabelEncoder()
bike_data['LOCATION_TYPE'] = le.fit_transform(bike_data['LOCATION_TYPE'])

# Scale numerical variables: 'BIKE_COST', 'month', and 'OCC_HOUR'

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(bike_data[features], bike_data['Theft_Risk'], test_size=0.3, random_state=42)

# Train a RandomForest Regressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predict probabilities and scale them to percentages
predictions = rf_model.predict(X_test) * 100  # Convert to percentage

# Evaluate the model
mse = mean_squared_error(y_test, predictions / 100)  # Convert back to original scale for MSE
r2 = r2_score(y_test, predictions / 100)

# Output results
print(f'Mean Squared Error: {mse:.4f}')
print(f'R-squared: {r2:.4f}')

# Display some predictions
result_df = X_test.copy()
result_df['Predicted_Theft_Risk (%)'] = predictions
print(result_df.head())
```

```python
from sklearn.cluster import KMeans
import numpy as np

coordinates = bike_data[['LAT_WGS84', 'LONG_WGS84']].dropna()

# Use KMeans to cluster coordinates into hotspots (e.g., 5 hotspots)
kmeans = KMeans(n_clusters=5, random_state=42)
bike_data['hotspot_cluster'] = kmeans.fit_predict(coordinates)

# Calculate hotspot centroids
hotspot_centroids = kmeans.cluster_centers_
```

```python
from scipy.spatial.distance import cdist

# Function to calculate distance from each point to nearest hotspot
def calculate_distance_to_hotspot(row, centroids):
    theft_location = np.array([[row['LAT_WGS84'], row['LONG_WGS84']]])
    distances = cdist(theft_location, centroids)
    return distances.min()  # Distance to nearest hotspot

# Apply distance calculation to each row
bike_data['distance_to_hotspot'] = bike_data.apply(lambda row: calculate_distance_to_hotspot(row, hotspot_centroids), axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Scale the distance to a range between 0 and 1
scaler = MinMaxScaler()
bike_data['normalized_distance'] = scaler.fit_transform(bike_data[['distance_to_hotspot']])
```

```python
features = bike_data[['normalized_distance', 'BIKE_COST', 'month', 'REPORT_HOUR']]  # Include other important features
target = bike_data['Theft_Risk']  # 1 for 'STOLEN', 0 for others

# Train/test split and Random Forest model as before
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

# Train the model
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
predictions = rf_model.predict(X_test) * 100  # Convert to percentage

# Evaluate the model
mse = mean_squared_error(y_test, predictions / 100)  # Convert back to original scale for MSE
r2 = r2_score(y_test, predictions / 100)

# Output results
print(f'Mean Squared Error: {mse:.4f}')
print(f'R-squared: {r2:.4f}')
```
